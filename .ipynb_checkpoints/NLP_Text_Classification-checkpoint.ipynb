## Предобработка данных

Удаление времени из столбца Даты:

```python
news_df['year'] = news_df['Date published'].dt.year
news_df['Date published'] = news_df['Date published'].dt.date
news_df['Date published']
```

```python
#Столбцы Author, Url, Keywords и т.д. могут содержать название категории / раздела, 
#это может повлиять на утечку данных при классификации
def fix_author(text):

    # Приводим к нижнему регистру
    text_clean = text.lower()

    # Удаляем категорию/раздел, чтобы избежать утечки данных
    remove_words = []
    remove_words.extend(news_df.Category.unique())
    remove_words.extend(news_df.Section.unique())
    # дополнительно удаляем слова
    remove_words.extend(['CNN','By'])
    remove_words = [word.lower() for word in remove_words]
    text_clean = ' '.join([word for word in text_clean.split() if word not in remove_words])
    
    # разделяем имена
    authors = re.split(',| and',text_clean)
    
    # удаляем специальные символы и цифры
    authors = [re.sub(r'[^a-z]', ' ', author) for author in authors]
    # соединяем через запятую
    authors = ", ".join(authors)

    return authors
```

```python
# Столбцы Article text, Headline и т.д. содержат необработанный текст (стоп-слова, строчные и заглавные буквы, спецсимволы,
#числа и т.д., это плохо отражается на восприятии текста, поэтому получим необработанные строк и вернем очищенные строки
def clean_text(web_text):

    # Приведение к нижнему регистру
    text_clean = web_text.lower()
    
    # Преобразование чисел в слова
    text_clean = ' '.join([num2words.num2words(i) if i.isdigit() else i for i in text_clean.split()])

    # Удаление специальных символов и цифр
    text_clean = re.sub(r'[^a-z]', ' ', text_clean)

    # Удаление стоп-слов, часто используемых слова в языке
    # Примеры стоп-слов в английском языке: “a”, “the”, “is”, “are” и т.д.
    stop_words = set(nltk.corpus.stopwords.words("english"))
    text_clean = ' '.join([word for word in text_clean.split() if word not in stop_words])

    # Лемматизация - это процесс объединения различных флективных форм слова
    # например, смотрела и смотреть - это одно и то же слово
    wordnet_lemmatizer = WordNetLemmatizer()
    lemmatized_list = []
    text_words = text_clean.split(" ")
     
    for word in text_words:
        lemmatized_list.append(wordnet_lemmatizer.lemmatize(word, pos="v"))

    text_clean = " ".join(lemmatized_list)
    
    # Стемминг - процесс приведения слов с изменяющимися при склонении или спряжении окончаний к их корню
    # формы, такие как сопоставление группы слов с одной и той же основой
    # даже если сама основа не является допустимым словом в Языке
    stemmer = PorterStemmer()
    stemmer.stem(text_clean)
     
    return text_clean
```

```python
# Применение функций к датасету
for index in news_df.index:
    
    text_author = news_df.loc[index,'Author']
    news_df.loc[index,'Author'] = fix_author(text_author)
    
    text_headline = news_df.loc[index,'Headline']
    news_df.loc[index,'Headline'] = clean_text(text_headline)
    
    text_headline2 = news_df.loc[index,'Second headline']
    news_df.loc[index,'Second headline'] = clean_text(text_headline2)
    
    text_description = news_df.loc[index,'Description']
    news_df.loc[index,'Description'] = clean_text(text_description)
    
    text_article = news_df.loc[index,'Article text']
    news_df.loc[index,'Article text'] = clean_text(text_article)
```

Удалим новости, попадающие под редкие категории, а также новости, которые были опубликованы в 2011 году. Таких значений очень мало для того, чтобы модель смогла обучиться на них:

```python
news_df = news_df[news_df.Category != 'vr']
news_df = news_df[news_df.Category != 'style']
news_df = news_df[news_df.Category != 'travel']
news_df = news_df[news_df.year != '2011']
news_df.shape
```

Разделение данных на обучающую и тестовую выборки:

```python
X_train, X_test, y_train, y_test = train_test_split(news_df["Article text"], news_df["Category_code"], 
                                                    test_size=0.2, random_state=True, stratify=Y)
```

Преобразование коллекции необработанных данных в матрицу функций TF-IDF (слова представляются в виде веторов):

```python
ngram_range = (1,2)
min_df = 10
max_df = 1.
max_features = 300

tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=ngram_range,
                        stop_words=None,
                        lowercase=False,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2',
                        sublinear_tf=True)

# Преобразование данных в матрицу в обучающей выборке
features_train = tfidf.fit_transform(X_train).toarray()
labels_train = y_train
print(features_train.shape)

# Преобразование данныхв матрицу в тестовой выборке
features_test = tfidf.transform(X_test).toarray()
labels_test = y_test
print(features_test.shape)
```

## Обучение моделей

##### Метод опорных векторов (SVM)

```python
start_time = time.time()
svc = svm.SVC(random_state=8)
svc.fit(features_train, labels_train)
svc_predict = svc.predict(features_test)
print("Accuracy: ", accuracy_score(labels_train, svc.predict(features_train)))
print("Метрики модели: ")
print(classification_report(labels_test, svc_predict))
print("Время выполнения", (time.time()-start_time))
```

```python
aux_df = news_df[['Category', 'Category_code']].drop_duplicates().sort_values('Category_code')
conf_matrix = confusion_matrix(labels_test, svc_predict)
plt.figure(figsize=(12.8, 6))
sns.heatmap(conf_matrix, 
            annot=True,
            xticklabels=aux_df['Category'].values, 
            yticklabels=aux_df['Category'].values,
            cmap="Reds")
plt.ylabel('Predicted') # Поменять оси местами
plt.xlabel('Actual')
plt.title('Confusion matrix')
plt.show()
```
